```python
#get list of specifications
f=open("specifications.txt")
spe_list=f.readlines()
f.close()
for i in range(len(spe_list)):
    spe_list[i]=spe_list[i][:-1]
big_spe=[]
small_spe=[]
for i in spe_list:
    t=i.split(',')
    big_spe.append(t[0])
    small_spe.append(t[1:])
print(big_spe)
print(small_spe)
```

```python
#class phone, use to get data and write data
class phone:
    def __init__(self,big_spe,small_spe):
        self.speci={}
        for i in range(len(big_spe)):
            k=big_spe[i]
            self.speci[k]={}
            for j in small_spe[i]:
                self.speci[k][j]=["NaN"]
    def add(self,big_spe,small_spe,val):
        if big_spe in self.speci.keys() and small_spe in self.speci[big_spe].keys():
            self.speci[big_spe][small_spe]=val
    def show(self):
        for keys,values in self.speci.items():
            print(keys)
            print(values)
    def write(self,f):
        s=""
        for key,value in self.speci.items():
            for k,v in value.items():
                t=""
                for vs in range(len(v)):
                    t+=v[vs]+";"
                t=t[:-1]
                #print(t)
                s+=t+","
        s=s[:-1]
        #print(s)
        f.write(s)
        f.write('\n')
    def write_key(self,f):
        s=""
        for key,value in self.speci.items():
            for k,v in value.items():
                s+=k
                s+=","
        s=s[:-1]
        f.write(s)
        f.write('\n')
```
