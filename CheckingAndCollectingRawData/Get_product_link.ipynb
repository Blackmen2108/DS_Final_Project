```python
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from requests_html import HTMLSession, HTML
```

```python
links = []
session = HTMLSession()
r = session.get('https://www.devicespecifications.com/en/brand-more')
brands = r.html.find('.brand-listing-container-news a')
for brand in brands:
    links.append(brand.attrs["href"])
links
```

```python
file = open("products_link.txt", "w", encoding='utf-8')
for link in links:
    r1 = session.get(link)
    r2 = r1.html.find('.model-listing-container-80 h3 a')
    for an in r2:
        file.write(str(an.attrs['href']) + "\n")
    session.close() 
    session = HTMLSession()
file.close()
```
