```python
import numpy as np
import pandas as pd
import math
import os
```

```python
files = os.listdir('C:/Users/DELL/Desktop/HK1_2019-2020/DataScience/FinalProject/DS_Final_Project/data/')
```

```python
mobile_df = pd.read_csv('C:/Users/DELL/Desktop/HK1_2019-2020/DataScience/FinalProject/DS_Final_Project/data/'+ files[0], encoding='utf-8', header=0)
for i in range(len(files)-1):
    temp = pd.read_csv('C:/Users/DELL/Desktop/HK1_2019-2020/DataScience/FinalProject/DS_Final_Project/data/' + files[i+1], encoding='utf-8', header=0)
    mobile_df = mobile_df.append(temp, ignore_index=True)
    del temp
```

```python
mobile_df
```

```python
# Xóa những cột không có price
nan_index = []
for i in range (mobile_df.shape[0]):
    if(type(mobile_df.iloc[i]['Price']) == float):
        if(math.isnan(mobile_df.iloc[i]['Price']) == True):
            nan_index.append(i)
mobile_df.drop(mobile_df.index[nan_index], inplace=True)
```

```python
# Xóa nhưng cột có tỉ lệ giá trị null trên 20%
ap = mobile_df.isnull().sum().values / mobile_df.shape[0]
del_col_index = []
for i in range(len(ap)):
    if(ap[i] >= 0.2):
        del_col_index.append(i)

mobile_df.drop(mobile_df.columns[del_col_index], axis=1, inplace=True )
```
