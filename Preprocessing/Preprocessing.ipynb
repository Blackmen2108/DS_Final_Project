```python
import numpy as np
import pandas as pd
```

```python
mobile_df = pd.read_csv('C:/Users/DELL/Desktop/HK1_2019-2020/DataScience/FinalProject/DS_Final_Project/CheckingAndCollectingRawData/dataset.csv', 
                        encoding='utf-8', header=0, sep ='\t')
```

```python
mobile_df.drop(['Brand','Name'], axis=1, inplace=True )
count_nan = []
del_col_index = []

for i in mobile_df.columns:
    count_nan.append(len(mobile_df[mobile_df[i] == 'No info'].index) / mobile_df.shape[0])

for i in range(len(count_nan)):
    if(count_nan[i] >= 0.2):
        del_col_index.append(i)

mobile_df.drop(mobile_df.columns[del_col_index], axis=1, inplace=True )
```

```python
# Xử lý cho bluetooth
Blue = []
for i in range(len(mobile_df)):
    if(mobile_df.iloc[i]['Blue'] != 'No info'):
        Blue.append(1)
    else:
        Blue.append(0)
        
mobile_df.drop(['Blue'], axis=1, inplace=True )
mobile_df['Bluetooth'] = Blue
```

```python
# Xử lý cho Dim
Width = []
Height = []
Thick = []

for i in range(len(mobile_df)):
    if (mobile_df.iloc[i]['Dim'] != 'No info'):
        temp = mobile_df.iloc[i]['Dim'].replace('mm', '').split(' x ')
        Width.append(float(temp[0]))
        Height.append(float(temp[1]))
        Thick.append(float(temp[2]))
    else:
        Width.append(float('nan'))
        Height.append(float('nan'))
        Thick.append(float('nan'))

mobile_df['Width'] = Width
mobile_df['Height'] = Height
mobile_df['Thick'] = Thick
mobile_df.drop(['Dim'], axis=1, inplace=True )
```

```python
# Xử lý cho Weight
Weight = []
for i in range(len(mobile_df)):
    if (mobile_df.iloc[i]['Weight'] != 'No info'):
        Weight.append(float(mobile_df.iloc[i]['Weight'].split(' ')[0]))
    else:
        Weight.append(float('nan'))

mobile_df.drop(['Weight'], axis=1, inplace=True )
mobile_df['Weight'] = Weight
```

```python
# Xử lý cho Memory card
Mem_card = []
for i in range(len(mobile_df)):
    temp = mobile_df.iloc[i]['Memory card']
    if(temp != 'No info'):
        if('micro' in temp):
            Mem_card.append('micro')
        else:
            if('Nano' in temp):
                Mem_card.append('nano')
            else:
                Mem_card.append('normal')
    else:
        Mem_card.append('No info')
        
mobile_df.drop(['Memory card'], axis=1, inplace=True )
mobile_df['Memory card'] = Mem_card  
```

```python
# Xử lý cho OS
OS = []
for i in range(len(mobile_df)):
    if('Android' in mobile_df.iloc[i]['OS']):
        OS.append('Android')
    else:
        OS.append(mobile_df.iloc[i]['OS'].split(' ')[0])

mobile_df.drop(['OS'], axis=1, inplace=True )
mobile_df['OS'] = OS
```

```python
# Xử lý cho SoC
SoC = []
for i in range(len(mobile_df)):
    if(mobile_df.iloc[i]['SoC'] != 'No info'):
        SoC.append(mobile_df.iloc[i]['SoC'].split(' ')[0].lower())
    else:
        SoC.append('No info')

mobile_df.drop(['SoC'], axis=1, inplace=True )
mobile_df['SoC'] = SoC
```

```python
# Xử lý cho Wi-Fi
Wi_Fi = []
for i in range(len(mobile_df)):
    if(mobile_df.iloc[i]['Wi-Fi'] != 'No info'):
        Wi_Fi.append(1)
    else:
        Wi_Fi.append(0)
        
mobile_df.drop(['Wi-Fi'], axis=1, inplace=True )
mobile_df['Wi-Fi'] = Wi_Fi
```

```python
# Xử lý cho Storage
#Storage = []
#for i in range(len(mobile_df)):
#    temp = []
#    if(mobile_df.iloc[i]['Storage'] != 'No info'):
#        store = mobile_df.iloc[i]['Storage'].split(' ')
#        if('MB' in store):
#            temp.append(float(store[0]) / 1024)
#        else:
#            for j in store:
#                if('GB' not in j):
#                    temp.append(float(j))
#        Storage.append(temp)
#    else:
#        Storage.append(float('nan'))

#mobile_df.drop(['Storage'], axis=1, inplace=True )
#mobile_df['Storage'] = Storage
```

```python
# Xử lý cho GPU
GPU_core = []

for i in range(len(mobile_df)):
    core = mobile_df.iloc[i]['GPU'].split(', ')
    # len = 1
    if(len(core) == 1):
        GPU_core.append('No info')
        
    # len = 2
    if(len(core) == 2):
        if('Cores' in core[1]):
            GPU_core.append(core[1][-1])
        else:
            GPU_core.append('No info')
    
    # len = 3
    if(len(core) == 3):
        GPU_core.append(core[2][-1])

mobile_df['GPU core'] = GPU_core
mobile_df.drop(['GPU'], axis=1, inplace=True )
```

```python
# Xử lý cho CPU
CPU_core = []

for i in range(len(mobile_df)):
    cpu = mobile_df.iloc[i]['CPU'].split(', ')
    if('Cores' in  cpu[-1]):
        CPU_core.append(cpu[-1][-1])
    else:
        CPU_core.append('No info')

mobile_df['CPU core'] = CPU_core
mobile_df.drop(['CPU'], axis=1, inplace=True )
```

```python
# Xử lý cho SIM
SIM = []
for i in range(len(mobile_df)):
    if('Nano-SIM' in mobile_df.iloc[i]['SIM']):
        SIM.append('Nano')
    else:
        if('Micro-SIM' in mobile_df.iloc[i]['SIM']):
            SIM.append('Micro')
        else:
            if('Mini-SIM' in mobile_df.iloc[i]['SIM']):
                SIM.append('Mini')
    if(mobile_df.iloc[i]['SIM'] == 'No info'):
        SIM.append('No info')

mobile_df.drop(['SIM'], axis=1, inplace=True )
mobile_df['SIM type'] = SIM
```

```python
# Xử lý cho số lượng SIM và USB
import os
SIM_card_num = []
USB_type = []
USB_version = []
k = 0
files = os.listdir('C:/Users/DELL/Desktop/HK1_2019-2020/DataScience/FinalProject/DS_Final_Project/data/')
for i in range(1,len(files)):
    temp = pd.read_csv('C:/Users/DELL/Desktop/HK1_2019-2020/DataScience/FinalProject/DS_Final_Project/data/' + files[i], encoding='utf-8', header=0)
    for j in range(len(temp)):
        if(type(temp.iloc[j]['Price']) != float):
            if(temp.iloc[j]['SIM card_Number of SIM cards'] != np.float64('NaN')):
                SIM_card_num.append(str(temp.iloc[j]['SIM card_Number of SIM cards']) + ' cards')
            else:
                SIM_card_num.append('No info')
            
            if(type(temp.iloc[j]['USB_Connector type']) != float):
                USB_type.append(temp.iloc[j]['USB_Connector type'])
            else:
                USB_type.append('No info')
                
            if(temp.iloc[j]['USB_Version'] != np.float64('NaN')):
                USB_version.append('version ' + str(temp.iloc[j]['USB_Version']))
            else:
                USB_version.append('No info')
            k +=1
    print(k)

mobile_df['Number of SIM cards'] = SIM_card_num
mobile_df['USB type'] = USB_type 
mobile_df['USB version'] = USB_version
mobile_df.drop(['USB'], axis=1, inplace=True )
```

```python
# Xử lý cho Pos
Pos = []
for i in range(len(mobile_df)):
    if(mobile_df.iloc[i]['Pos'] == 'No info'):
        Pos.append(0)
    else:
        Pos.append(1)
        
mobile_df.drop(['Pos'], axis=1, inplace=True )
mobile_df['Position tracking'] = Pos
```

```python
# Xử lý cho Dis
Dis_size = []
Dis_res = []
Dis_color = []
color = False

for i in range(len(mobile_df)):
    res = False
    color = False
    temp = mobile_df.iloc[i]['Dis'].split(', ')
    Dis_size.append(float(temp[0].split(' ')[0]))
    for j in temp:
        if(' bit' in j):
            Dis_color.append(j.split(' ')[0])
            color = True
        if('pixels' in j):
            temp_1 = j.split(' ')
            Dis_res.append(round(float(temp_1[0]) * float(temp_1[2]) / 1000000, 1))
    if (color == False):
        Dis_color.append('No info')

mobile_df['Display size'] = Dis_size
mobile_df['Display resolution'] = Dis_res
mobile_df['Display color depth'] = Dis_color
mobile_df.drop(['Dis'], axis=1, inplace=True )
```

```python
# Xử lý cho Camera
Im_res = []
Vid_res = []
FPS = []
for i in range(len(mobile_df)):
    if(mobile_df.iloc[i]['Cam'] == 'No info'):
        Im_res.append(float('nan'))
        Vid_res.append(float('nan'))
        FPS.append(float('nan'))
    else:
        temp = mobile_df.iloc[i]['Cam'].split(', ')
        if(len(temp) == 1):
            temp_1 = temp[0].split(' ')
            Im_res.append(round(float(temp_1[0]) * float(temp_1[2]) / 1000000, 1))
            Vid_res.append(float('nan'))
            FPS.append(float('nan'))
            
        if(len(temp) == 3):
            temp_1 = temp[0].split(' ')
            temp_2 = temp[1].split(' ')
            temp_3 = temp[2].split(' ')
            Im_res.append(round(float(temp_1[0]) * float(temp_1[2]) / 1000000, 1))
            Vid_res.append(round(float(temp_2[0]) * float(temp_2[2]) / 1000000, 1))
            FPS.append(int(temp_3[0]))
            
        if(len(temp) == 2):
            temp_1 = temp[0].split(' ')
            Im_res.append(round(float(temp_1[0]) * float(temp_1[2]) / 1000000, 1))
            if('pixels' in temp[1]):
                temp_2 = temp[1].split(' ')
                Vid_res.append(round(float(temp_2[0]) * float(temp_2[2]) / 1000000, 1))
                FPS.append(float('nan'))
            if('fps' in temp[1]):
                temp_3 = temp[1].split(' ')
                FPS.append(int(temp_3[0]))
                Vid_res.append(float('nan'))

mobile_df['Image resolution'] = Im_res
mobile_df['Video resolution'] = Vid_res
mobile_df['FPS'] = FPS
mobile_df.drop(['Cam'], axis=1, inplace=True )
```

```python
# Xử lý cho Battery
Bat_type = []
Bat_power = []

for i in range(len(mobile_df)):
    temp = mobile_df.iloc[i]['Bat'].split(', ')
    if(len(temp) == 2):
        Bat_type.append(temp[1])
        Bat_power.append(float(temp[0].split(' ')[0]))
    else:
        if('mAh' in temp[0]):
            Bat_power.append(float(temp[0].split(' ')[0]))
            Bat_type.append('No info')
        else:
            Bat_power.append(float('nan'))
            Bat_type.append(temp[0])

mobile_df['Battery type'] = Bat_type
mobile_df['Battery power'] = Bat_power
mobile_df.drop(['Bat'], axis=1, inplace=True )
```

```python
# Xử lý cho RAM, Storage và Price
RAM = []
Storage = []
Price = []

for i in range(len(mobile_df)):
    if(mobile_df.iloc[i]['RAM'] != 'No info'):
        temp = mobile_df.iloc[i]['RAM'].split(', ')
        if('MB' in temp[(len(temp)-1) // 2]):
            ram_temp = round(float(temp[(len(temp)-1) // 2].split(' ')[0]) / 1024, 2)
            #RAM.append(str(ram_temp) + ' GB') 
            #sửa ở đây
            RAM.append(ram_temp)
        else:
            #RAM.append(temp[(len(temp)-1) // 2]) 
            #sửa ở đây
            RAM.append(float(temp[(len(temp)-1) // 2].split(' ')[0]))
    else:
        #RAM.append('No info') 
        #sửa ở đây
        RAM.append(float('nan'))
        
    if(mobile_df.iloc[i]['Storage'] != 'No info'):
        store = mobile_df.iloc[i]['Storage'].split(', ')
        if('MB' in store[(len(store)-1) // 2]):
            store_temp = round(float(store[(len(store)-1) // 2].split(' ')[0]) / 1024, 2)
            #Storage.append(str(store_temp) + ' GB') 
            #sửa ở dây
            Storage.append(store_temp)
        else:
            #Storage.append(store[(len(store)-1) // 2]) 
            #sửa ở đây
            Storage.append(float(store[(len(store)-1) // 2].split(' ')[0]))
    else:
        #Storage.append('No info') 
        #sửa ở đây
        Storage.append(float('nan'))
        
    price = mobile_df.iloc[i]['Average price'].replace('[', '').replace(']', '').replace("'", '').split(', ')
    Price.append(float(price[(len(price)-1) // 2]))

mobile_df.drop(['RAM'], axis=1, inplace=True)
mobile_df.drop(['Storage'], axis=1, inplace=True)
mobile_df.drop(['Average price'], axis=1, inplace=True)
mobile_df['RAM'] = RAM
mobile_df['Storage'] = Storage
```

```python
# Xử lý cho Price
Range_price = []
for i in range(len(Price)):
    # giá rẻ
    if(Price[i] < 300):
        Range_price.append(0)
    # giá trung bình
    if(Price[i] >= 300 and Price[i] < 600):
        Range_price.append(1)
    # giá cao
    if(Price[i] >= 600 and Price[i] < 900):
        Range_price.append(2)
    # giá rất cao
    if(Price[i] >= 900):
        Range_price.append(3)

mobile_df['Price range'] = Range_price
```
