```python
from requests_html import HTMLSession, HTML
import time

session = HTMLSession()
price_of_product = []
name_of_product = []
rf = open('link_to_price.txt', 'r')
i = 1
link = rf.readline()
while(link != ''):
    print(i)
    product_prices = []
    r = session.get(link.strip())
    name = r.html.find('.section-header h2', first = True).text
    prices = r.html.find('.price')
    for price in prices:
        product_prices.append(price.text)
    price_of_product.append(product_prices)
    name_of_product.append(name)
        
    print(name)
    print(product_prices)
    
    i = i+1
    session.close() 
    session = HTMLSession()
    link = rf.readline()
    time.sleep(1)
```

```python
wf = open('name_and_price.txt', 'w')
for i in range(len(price_of_product)):
    wf.write('name' + '\t' + 'prices' + '\n')
    wf.write(name_of_product[i] + '\t')
    for x in range(len(price_of_product[i])):
        if(len(price_of_product[i]) == 1):
            wf.write(price_of_product[i][0])
        else:
            if(x == (len(price_of_product[i])-1)):
                wf.write(price_of_product[i][-1])
            else:
                wf.write(price_of_product[i][x] + ', ')
    wf.write('\n\n')
wf.close()
```
