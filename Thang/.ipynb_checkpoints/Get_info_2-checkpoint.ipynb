```python
from requests_html import HTMLSession, HTML
import requests
import time

rf = open("products_links.txt", "r")
wf = open("price_links.txt", "w")
url = "https://www.devicespecifications.com/en/price/"
session = HTMLSession()
line = rf.readline()
has_price = False

while (line != ""):
    r = session.get(line.strip())
    for tab in r.html.find('.tab-container a'):
        if (tab.text.find("Prices") != -1):
            has_price = True;
    
    if(has_price == True):
        code = line.strip().split("/")[-1]
        price_link = url + code
        print(price_link)
        wf.write(price_link + "\n")
    
    has_price = False
    line = rf.readline()
    session.close() 
    session = HTMLSession()
    time.sleep(1)

rf.close()
wf.close()
```
