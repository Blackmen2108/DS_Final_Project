```python
import pandas as pd
from decimal import Decimal
```

```python
label = set()
rf = open('information.txt', 'r')
i = 1
line = rf.readline()
while(line != ''):
    if((i%3) == 1):
        labs = line.strip().split('\t')
        for lab in labs:
            label.add(lab)
    
    i = i+1
    line = rf.readline()
rf.close()
```

```python
Battery = []
Bluetooth = []
CPU = []
Camera = []
Cores = []
Dimensions = []
Display = []
GPU = []
Memory_cards = []
OS = []
Positioning = []
RAM = []
SIM_card = []
SoC = []
Storage = []
USB = []
Weight = []
Wi_Fi = []

rf = open('information.txt', 'r')
for m in range(i):
    line = rf.readline()
    if((m%3) == 0):
        labs = line.strip().split('\t')
    if((m%3) == 1):
        detail = line.strip().split('\t')
    if((m%3) == 2):
        if('Battery' in labs):
            index = labs.index('Battery')
            Battery.append(detail[index])
        else:
            Battery.append('No info')
    
        if('Bluetooth' in labs):
            index = labs.index('Bluetooth')
            Bluetooth.append(detail[index])
        else:
            Bluetooth.append('No info')
        
        if('CPU' in labs):
            index = labs.index('CPU')
            CPU.append(detail[index])
        else:
            CPU.append('No info')
        
        if('Camera' in labs):
            index = labs.index('Camera')
            Camera.append(detail[index])
        else:
            Camera.append('No info')
        
        if('Cores' in labs):
            index = labs.index('Cores')
            Cores.append(detail[index])
        else:
            Cores.append('No info')
        
        if('Dimensions' in labs):
            index = labs.index('Dimensions')
            Dimensions.append(detail[index])
        else:
            Dimensions.append('No info')
    
        if('Display' in labs):
            index = labs.index('Display')
            Display.append(detail[index])
        else:
            Display.append('No info')
    
        if('GPU' in labs):
            index = labs.index('GPU')
            GPU.append(detail[index])
        else:
            GPU.append('No info')
    
        if('Memory cards' in labs):
            index = labs.index('Memory cards')
            Memory_cards.append(detail[index])
        else:
            Memory_cards.append('No info')
    
        if('OS' in labs):
            index = labs.index('OS')
            OS.append(detail[index])
        else:
            OS.append('No info')
    
        if('Positioning' in labs):
            index = labs.index('Positioning')
            Positioning.append(detail[index])
        else:
            Positioning.append('No info')
    
        if('RAM' in labs):
            index = labs.index('RAM')
            RAM.append(detail[index])
        else:
            RAM.append('No info')
    
        if('SIM card' in labs):
            index = labs.index('SIM card')
            SIM_card.append(detail[index])
        else:
            SIM_card.append('No info')
        
        if('SoC' in labs):
            index = labs.index('SoC')
            SoC.append(detail[index])
        else:
            SoC.append('No info')
        
        if('Storage' in labs):
            index = labs.index('Storage')
            Storage.append(detail[index])
        else:
            Storage.append('No info')
        
        if('USB' in labs):
            index = labs.index('USB')
            USB.append(detail[index])
        else:
            USB.append('No info')
        
        if('Weight' in labs):
            index = labs.index('Weight')
            Weight.append(detail[index])
        else:
            Weight.append('No info')
        
        if('Wi-Fi' in labs):
            index = labs.index('Wi-Fi')
            Wi_Fi.append(detail[index])
        else:
            Wi_Fi.append('No info')
```

```python
brand = []
name = []
avg_price = []
fo = open('name_and_price.txt', 'r')
for m in range(i):
    line = fo.readline()
    if((m%3) == 1):
        average = 0
        info = line.strip().split('\t')
        
        brand_name = info[0].split(' ')
        brand.append(brand_name[-2])
        name.append(brand_name[-1])
        
        prices = info[1].replace('$', '')
        prices = prices.split(', ')
        for price in prices:
            average += Decimal(price)
        average = average/len(prices)
        avg_price.append(round(average, 2))
```

```python
dict = {'Brand': brand, 'Name': name, 'Dim': Dimensions, 'Weight': Weight, 'Bat': Battery, 'Blue': Bluetooth, 'CPU': CPU, 
        'Cam': Camera, 'Cores': Cores, 'Dis': Display, 'GPU': GPU, 'Memory card': Memory_cards, 'OS': OS, 'Pos': Positioning, 
        'RAM': RAM, 'SIM': SIM_card, 'SoC': SoC, 'Storage': Storage, 'USB': USB, 'Wi-Fi': Wi_Fi, 'Average price': avg_price}
df = pd.DataFrame(dict)
df
```

```python
df.to_csv (r'dataset.csv', index = None, header=True, sep = '\t')
```
